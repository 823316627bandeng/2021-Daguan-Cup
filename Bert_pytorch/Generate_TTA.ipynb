```python
import pandas as pd
import numpy as np
from gensim.models import Word2Vec
import pandas as pd
import jieba
import os
```

```python
#train = pd.read_csv('datagrand_2021_train.csv')
test = pd.read_csv('datagrand_2021_test.csv')

```

```python
#对测试集进行tta
#如果是对训练集进行tta，就把下面这句删除
train=test
```

```python
#先统计每个符号出现的位置
#然后再将所有符号换成英文逗号
#最后再根据逗号进行还原
fuhao=['，','！','。','？']
tmp=train.text.tolist()
```

```python
totalFuhao=[]
for text in tmp:
    tF=[]
    t=text.split()
    for j in t:
        if j in fuhao:
            tF.append(j)
            # print("ok")
    totalFuhao.append(tF)
```

```python
print(len(totalFuhao))
```

```python
def getClean(document):
    text = str(document)
    text = text.replace('，', ',')
    text = text.replace('！', ',')
    text = text.replace('？', ',')
    text = text.replace('。', ',')
    return text
```

```python
def suffer(document):
    text=str(document)
    t=text.split(',')
    newT=t[::-1]
    return " , ".join(newT)
```

```python
train['text']=train['text'].apply(lambda x:getClean(x))
```

```python
#句子逆序
train['text']=train['text'].apply(lambda x: suffer(x))
```

```python
#符号还原
def tranform(df):
    ixd=0
    totaldx=0
    ans=[]
    for text in df:
        arr=[]
        dinx=0
        t=text.split()
        if ixd==0:
            print(t)
        for j in t:
            if j==',':
                arr.append(totalFuhao[totaldx][dinx])
                dinx+=1
            else :
                arr.append(j)
        ixd+=1
        totaldx+=1
        if ixd==1 :
            print(" ".join(arr))
        ans.append(" ".join(arr))
    return ans
    
```

```python
#将倒序后的句子进行符号还原
newText=train['text'].tolist()
neT=tranform(newText)
```

```python
train['text']=neT
```

```python
train.to_csv("./ttatest.csv",index=False)
```
